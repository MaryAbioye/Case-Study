```python
import pandas as pd

# Step 1: Load the Value and Volume Dataset
my_data = "Data Test.xlsx"
value_df = pd.read_excel(my_data, sheet_name="Value", header=11)
volume_df = pd.read_excel(my_data, sheet_name="Volume", header=11)

# show few rows
value_df.head()
volume_df.head()
```

```python
# Keep only relevant columns
value_cols = [col for col in value_df.columns if isinstance(col, str) and col.startswith("w/e")]
volume_cols = [col for col in volume_df.columns if isinstance(col, str) and col.startswith("w/e")]

value_df = value_df[["Price Sector", "Reportable Owner", "Brand House"] + value_cols]
volume_df = volume_df[["Price Sector", "Reportable Owner", "Brand House"] + volume_cols]

value_df.head()
```

```python
# Reshape value and volume data from wide to long format
value_long = pd.melt(
    value_df,
    id_vars=["Price Sector", "Reportable Owner", "Brand House"],
    var_name="Date",
    value_name="Value"
)

volume_long = pd.melt(
    volume_df,
    id_vars=["Price Sector", "Reportable Owner", "Brand House"],
    var_name="Date",
    value_name="Volume"
)

value_long.head()


```

```python
# Remove 'w/e ' prefix and convert to datetime
value_long["Date"] = pd.to_datetime(value_long["Date"].str.replace("w/e ", ""), dayfirst=True)
volume_long["Date"] = pd.to_datetime(volume_long["Date"].str.replace("w/e ", ""), dayfirst=True)

value_long["Date"].head()

```

```python
# Merge value and volume data
merged_data = pd.merge(
    value_long,
    volume_long,
    on=["Price Sector", "Reportable Owner", "Brand House", "Date"],
    how="inner"
)

merged_data.head()
```

```python
# Create Holding_Company column by replacing ITUK with IMB
merged_data["Holding_Company"] = merged_data["Reportable Owner"].replace("ITUK", "IMB")

# check column
merged_data[["Reportable Owner", "Holding_Company"]].drop_duplicates().head()
```

```python
# Create PS_HC_BH_Group column
merged_data["PS_HC_BH_Group"] = (
    merged_data["Price Sector"] + "_" +
    merged_data["Holding_Company"] + "_" +
    merged_data["Brand House"]
)

# check column
merged_data[["PS_HC_BH_Group"]].head()
```

```python
# Calculate total value and volume per Date
merged_data["Value_T"] = merged_data.groupby("Date")["Value"].transform("sum")
merged_data["Volume_T"] = merged_data.groupby("Date")["Volume"].transform("sum")
```

```python
# Calculate share columns
merged_data["Value_Share"] = merged_data["Value"] / merged_data["Value_T"]
merged_data["Volume_Share"] = merged_data["Volume"] / merged_data["Volume_T"]

# check columns
merged_data[[
    "Date", "Price Sector", "Holding_Company", "Brand House","PS_HC_BH_Group",
    "Value", "Volume", "Value_T", "Volume_T", "Value_Share", "Volume_Share"
]].head()
```

```python
# Step 10: Final answer filtering
final_output = merged_data[
    (merged_data["Price Sector"] == "Premium") &
    (merged_data["Holding_Company"] == "IMB") &
    (merged_data["Brand House"] == "JPS") &
    (merged_data["Date"] == pd.to_datetime("2024-04-21"))
][["Date", "Value_Share", "Volume_Share"]]

# Show the result
final_output
```
